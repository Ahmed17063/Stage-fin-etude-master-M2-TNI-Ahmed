```python
import os
import openpyxl
import pandas as pd
import re
import cv2
import matplotlib.pyplot as plt
```

Création de la table pandas des images

```python
CurrFold=r"N:\tribologie_numerique\sources\Images"

source1, source2, source3="te", "profilometre", "video"
type1, type2="profil", "graphe"
materiau1,materiau2= "fixe", "mobile"

nom_images=[a for a in os.listdir(CurrFold) if not a.endswith(".db")]

chemin_images=[os.path.join(CurrFold,f) for f in nom_images]

source_images=[0 for a in nom_images]
for i in range(len(nom_images)):
    x=nom_images[i].split("_")
    if source1 in x[1].lower():
        source_images[i]=source1
    elif source2 in x[1].lower():
        source_images[i]=source2
    elif source3 in x[1].lower():
        source_images[i]=source3

type_materiau=[0 for a in nom_images]
for k in range(len(nom_images)):
    x=nom_images[k].split("_")
    if materiau1 in x[4].lower():
        source_images[i]=materiau1
    elif materiau2 in x[4].lower():
        source_images[i]=materiau2

type_image=[0 for a in nom_images]
for i in range(len(nom_images)):
    if type1 in nom_images[i].lower():
        type_image[i]=type1
    elif type2 in nom_images[i].lower():
        type_image[i]=type2

essai=[0 for a in nom_images]
for k in range(len(nom_images)):
   x=nom_images[k].split("_")
   essai[k]=x[0]+"_"+x[1]+"_"+x[2]
    
image_data= {"nom":nom_images,"source": source_images,"type_mat" : type_materiau, "type_image":type_image ,"chemin":chemin_images, "essai":essai}
image_df=pd.DataFrame(data=image_data)
image_df
```

Enregistrement de la table dans le fichier "images.csv"

```python
#save our pandas dataframe in images.csv (write dataframe to csv file with default params)
image_df.to_csv(r"N:\tribologie_numerique\sources\images.csv")
```
